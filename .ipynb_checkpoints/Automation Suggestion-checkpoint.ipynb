```python
# TO strip at the start and end '*'
automation_df['Keyword']=automation_df['Keyword'].apply(lambda x: x.strip(punctuation))
```

```python
automation_df['Keyword']=automation_df['Keyword'].str.replace('*',' ')
```

```python
# To remove punchuation, numbers and converting the string to lower
automation_df['Keyword']=automation_df['Keyword'].str.lower()
```

```python
automation_df['Keyword']=automation_df['Keyword'].apply(lambda row: " ".join([w.lemma_ for w in nlp(row)]))
```

```python

# importing pandas as pd 
import pandas as pd 
    
# dictionary of lists 
temp_dict = {'two':["health check", "hold space", "hourly check", "need no"], 
        'three': ["high processor load", "virtual machine", "mail box need", "need install software"], 
        } 
   
# creating a dataframe from a dictionary  
df = pd.DataFrame(temp_dict) 
df.head()
  
```

```python
# columns = list(df) 
```

```python
automation_col=[]
# for i in columns:
for record in range(0,len(df)):
    
    grams_list_record=[]
    grams_list_record.append(df['two'][record])
    grams_list_record.append(df['three'][record])
    n = {k: dict_keyword[k] for k in grams_list_record if k in dict_keyword}
    
    flag=bool(n)
    res=''
    if flag==True:
        res=max(n, key=n.get)
    else:
        res=''
    print(n)
    automation_col.append(res)
```

```python
automation_col
```

```python
n = {k: dict_automation_text[k] for k in automation_col }
```

```python
automation_text=list(n.values())
```

```python
automation_text
```

```python
ticket_topgrams_df['Automation Category']=pd.DataFrame({'Automation Category':automation_text})
```
