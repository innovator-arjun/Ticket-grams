```python
import os
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from nltk.tokenize import RegexpTokenizer
import re

import spacy 
nlp=spacy.load('en_core_web_sm')
```

```python
#pointing the source and listing the file names under that path
path = os.chdir('C://Users//ar393556//Documents//Ticket grams//CIdata')
files = os.listdir(path)
files
```

```python
#Fetching only xlsx files
files_xls = [f for f in files if f[-4:] == 'xlsx']
files_xls
```

```python
df = pd.DataFrame()

#Reading the files and appending the data in the DataFrame format
for f in files_xls:
    data = pd.read_excel(f,sheet_name='CI-Data')
    df = df.append(data)
```

```python
#Taking only TicketID and Summary which are essential

ticket_df_full = pd.DataFrame(df, columns= ['Summary'])
```

```python
ticket_df=ticket_df_full['Summary']
```

```python
#no. of records and columns
ticket_df.shape
```

```python
#Dropping null/blank records
ticket_df=ticket_df.dropna()
```

```python
#New table data shape
ticket_df.shape
```

```python
ticket_df.sample(20)
```

```python
# To remove punchuation, numbers and converting the string to lower
ticket_df = ticket_df.str.lower().str.replace(r'[^a-z\s]', '')
```

```python
ticket_df_new = ticket_df[ticket_df.notnull()]
```

```python
nlp.Defaults.stop_words.add('thanks')
nlp.Defaults.stop_words.add('please')
nlp.Defaults.stop_words.add('team')
nlp.Defaults.stop_words.add('dear')
```

```python
# nlp.Defaults.stop_words.remove('not')
# nlp.Defaults.stop_words.remove('cannot')
# nlp.Defaults.stop_words.remove('nothing')
```

```python
stop_words=nlp.Defaults.stop_words
ticket_df_new = ticket_df_new.apply(lambda x: " ".join(x for x in x.split() if x not in stop_words))
```

```python
ticket_df_new.head()    
```

```python
freq = pd.Series(' '.join(ticket_df_new).split()).value_counts()
```

```python
imp_words=freq[freq<5]
```

```python
imp_words
```

```python
freq = list(imp_words.index)
ticket_df_new = ticket_df_new.apply(lambda x: " ".join(x for x in x.split() if x not in freq))
```

```python
ticket_df_new = ticket_df_new.apply(lambda x: " ".join(x for x in x.split() if len(x)>=3 and len(x)<12))
```

```python
ticket_df_new
```

```python
from gensim import corpora
tokenized_sents= ticket_df_new.values.tolist()
texts = [[i for i in doc.split()] for doc in tokenized_sents]

dictionary = corpora.Dictionary(texts)
```

```python
print(dictionary)
```

```python
for words in dictionary.values():
    print(words)
```

```python
import pandas as pd

(pd.DataFrame.from_dict(data=dictionary, orient='index')
   .to_csv('dict_file.csv', header=False))
```
